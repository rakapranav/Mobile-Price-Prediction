```python
data = data.rename(columns={"blue":"bluetooth",
                           "pc":"pc_megapixel",
                           "fc":"fc_megapixel",
                           "m_dep":"m_depth"})
```

```python
data.head()
```

```python
#to check the duplicated record
dupes = data.duplicated()
dupes.head()
sum(dupes)
```

```python
#to remove the duplicated records
data = data.drop_duplicates()
data.shape
```

```python
# to check if the column contains null values
data.isnull().sum()
```

```python
# Seperate the numerical data and categorical data
numeric_data = data.drop(["bluetooth","dual_sim","four_g","three_g","touch_screen","wifi","price_range"],axis=1)
```

```python
numeric_data.head()
```

```python
category_data = data[["bluetooth","dual_sim","four_g","three_g","touch_screen","wifi","price_range"]]
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Boxplot the ram column
fig, ax = plt.subplots(figsize = (10,8))
sns.boxplot(numeric_data['ram'],orient="v")
```
