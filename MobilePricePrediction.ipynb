```python
data = data.rename(columns={"blue":"bluetooth",
                           "pc":"pc_megapixel",
                           "fc":"fc_megapixel",
                           "m_dep":"m_depth"})
```

```python
data.head()
```

```python
#to check the duplicated record
dupes = data.duplicated()
dupes.head()
sum(dupes)
```

```python
#to remove the duplicated records
data = data.drop_duplicates()
data.shape
```

```python
# to check if the column contains null values
data.isnull().sum()
```

```python
# Seperate the numerical data and categorical data
numeric_data = data.drop(["bluetooth","dual_sim","four_g","three_g","touch_screen","wifi","price_range"],axis=1)
```

```python
numeric_data.head()
```

```python
category_data = data[["bluetooth","dual_sim","four_g","three_g","touch_screen","wifi","price_range"]]
```

```python
category_data.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Boxplot the ram column
fig, ax = plt.subplots(figsize = (10,8))
sns.boxplot(numeric_data['ram'],orient="v")
```

```python
fig, ax = plt.subplots(figsize = (10,8))
bp = sns.boxplot(data=numeric_data)
bp.set_xticklabels(bp.get_xticklabels(),rotation=90)
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_array = scaler.fit_transform(numeric_data)
```

```python
scaled_data = pd.DataFrame(scaled_array,columns=numeric_data.columns)
scaled_data.head()
```

```python
fig, ax = plt.subplots(figsize = (10,8))
bp = sns.boxplot(data=scaled_data)
bp.set_xticklabels(bp.get_xticklabels(),rotation=90)
```

```python
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier_removed_data = numeric_data[~ ((numeric_data < (Q1 - 1.5 * IQR)) | ((numeric_data > (Q3 + 1.5 * IQR)))).any(axis=1)]
```

```python
from sklearn.model_selection import train_test_split

scaled_data = scaled_data.reset_index()
category_data = category_data.reset_index()
```

```python
final_df = pd.concat([scaled_data,category_data], axis=1)
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
x = final_df.drop('price_range', axis=1)
y = final_df['price_range']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(solver="lbfgs",
                                   multi_class="multinomial",
                                   max_iter = 10000)
```

```python
logistic_model.fit(X_train,y_train)
```

```python
logistic_model.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
svm_model.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
```

```python
decision_tree.score(X_test,y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
names = ("LR","DecisionTree","SVM")
results = [logistic_model.score(X_test,y_test) * 100,decision_tree.score(X_test,y_test)*100,
          svm_model.score(X_test,y_test)*100]

```

```python
y_pos = np.arange(len(names))

plt.bar(y_pos, results,alpha=0.7)
plt.xticks(y_pos, names)
plt.ylabel('Accuracy')
plt.title('Classifier Result')

plt.show()
```
